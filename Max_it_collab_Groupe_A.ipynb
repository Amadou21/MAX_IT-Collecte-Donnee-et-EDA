## L'age et la frequence d'utilisation

```python
def recherche(row):
    if 'Plusieurs fois par jour' in row['9_freq_utilisation']:
        return 20
    elif 'Quotidiennement' in row['9_freq_utilisation']:
        return 15
    elif 'Hebdomadairement' in row['9_freq_utilisation']:
        return 10
    elif 'Mensuellement' in row['9_freq_utilisation']:
        return 5
    elif 'Rarement' in row['9_freq_utilisation']:
        return 0

df['9_freq_utilisation_code'] = df.apply(recherche, axis=1)
```

```python
#!pip install scipy
from scipy import stats

# Create a contingency table using the numerical variable
contingence_table = pd.crosstab(df['1_age'], df['9_freq_utilisation_code'])

# Perform the chi-square test with the contingency table
chi2, p_value, dof, expected = stats.chi2_contingency(contingence_table.values)


# Interpret the chi-square test results
if p_value < 0.05:
    print("Il existe une association statistiquement significative entre l'âge et la fréquence d'utilisation (p-value = {:.3f})".format(p_value))
else:
    print("Il n'y a pas d'association statistiquement significative entre l'âge et la fréquence d'utilisation (p-value = {:.3f})".format(p_value))

# Calculate and display the standardized residuals
standardized_residuals = (contingence_table.values - expected) / np.sqrt(expected)
print("\nRésidus standardisés:")
print(standardized_residuals)

# Visualize the relationship using a bar chart
df.groupby('1_age')['9_freq_utilisation_code'].count().plot(kind='bar')
plt.xlabel('Tranche d\'âge')
plt.ylabel('Fréquence d\'utilisation moyenne')
plt.title('Fréquence d\'utilisation moyenne par tranche d\'âge')
plt.show()
```
